```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('E:\\7th sem\\ML-LAB\\ML_lab_7th-Sem-main\\ML_lab_7th-Sem-main\\student_scores.csv')
```

```python
df
```

```python
df['x']=df['Hours']-df['Hours'].mean()
```

```python
df['y']=df['Scores']-df['Scores'].mean()
```

```python
df['x*x']=df['x']*df['x']
```

```python
df['y*y']=df['y']*df['y']
```

```python
df['x*y']=df['x']*df['y']
```

```python
df
```

```python
r=0
r=df['x*y'].sum()/np.sqrt(df['x*x'].sum()*df['y*y'].sum())
```

```python
r
```

```python
std_x=df['Hours'].std()
```

```python
std_y=df['Scores'].std()
```

```python
m=r*(std_y/std_x)
```

```python
m
```

```python
mean_x=df['Hours'].mean()
```

```python
mean_y=df['Scores'].mean()
```

```python
c=mean_y-m*mean_x
```

```python
c
```

```python
y_pred_reg=m*df['Hours']+c
df['y_pred']=y_pred_reg
```

```python
def ssr(pred):
    res=0
    for i in range(len(pred)):
        res=res+(pred[i]-pred.mean())**2
    return res
```

```python
def sst(actual):
    res=0
    for i in range(len(actual)):
        res=res+(actual[i]-actual.mean())**2
    return res
```

```python
def sse(actual,pred):
    res=0
    
    for i in range(len(actual)):
        res=res+(actual[i]-pred[i])**2
    return res
```

```python
print(ssr(df['y_pred']))
```

```python
print(sst(df['Scores']))
```

```python
print(sse(df['Scores'],df['y_pred']))
```
